```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('股东增减持.csv')
df
```

```python
df['评估日期'] = pd.DatetimeIndex(df['公告日']).year
pivot_table = df.pivot_table(index='公司简称', columns=['评估日期'], values='变动数量(万股)', aggfunc=np.sum)
pivot_table = pivot_table.fillna(0.0)
pivot_table
```

```python
pivot_table['total'] = pivot_table.apply(lambda x: x.sum(), axis=1)
pivot_table
```

```python
pivot_table.to_excel('pivot_table.xlsx')
```
