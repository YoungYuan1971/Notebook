```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('待加权平均.xlsx')
df
```

```python
df['number_proportion'] = df['number'] / df.groupby('name')['number'].transform('sum')
df  # 各品种占本品种总量的比重
```

```python
group = df.groupby('name').agg({'number': 'sum', 'price': 'mean'})
group = group.rename(
    columns={'number': 'sum_number', 'price': 'average_price'})
group
```

```python
group['weighted_mean_price'] = df.groupby('name').apply(lambda x: np.average(x['price'], weights=x['number']))
group
```
