```python
import pandas as pd
```

```python
datas = pd.read_csv('股东增减持.csv', dtype={'代码': 'object'})
datas
```

```python
group = datas.groupby(by='公司简称').agg(
    {'变动数量(万股)': 'sum', '占总股本比例': 'sum', '变动后持股总数(万股)': 'sum', \
     '变动后占总股本比例': 'mean', '涨跌幅': 'mean'})
group
```

```python
group.to_csv('分类汇总.csv', sep=',', encoding='utf-8-sig')
```
