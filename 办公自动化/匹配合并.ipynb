```python
import pandas as pd
```

```python
info = pd.read_excel('students.xlsx', sheet_name='info')
score = pd.read_excel('students.xlsx', sheet_name='score')
score
```

```python
match = pd.merge(info, score[['ID', 'score', 'ranking']], on='ID', how='left')
match
```

```python
match.sort_values(by=['ranking'], inplace=True)
match
```

##### 如果有同名的表存在，则先删除并保存

```python
with pd.ExcelWriter ('students.xlsx', engine='openpyxl',mode='a') as writer:
    wb = writer.book
    if 'match' in wb.sheetnames:
        wb.remove(wb['match'])
```

##### 添加新的表

```python
with pd.ExcelWriter('students.xlsx', engine='openpyxl', mode='a') as writer:
    match.to_excel(writer, sheet_name='match', index=False)
```
